```python
import numpy as np
import pandas as pd
```

```python
df_tot_emissions = pd.read_excel('TIMES_PanEU_total_emissions.xls',
                                header=[0,1],
                                index_col=[0,1],
                                sheet_name="Emission of CO2")
```

```python
df_captured_emissions = pd.read_excel('TIMES_PanEU_captured_emissions.xls',
                                header=[0,1],
                                index_col=[0,1],
                                sheet_name="Emission - Separated CO2") 
```

## Make pivot table from total emissions

```python
# get names of different indexes
regions   = df_tot_emissions.index.levels[0]
sectors   = df_tot_emissions.index.levels[1]
years     = df_tot_emissions.columns.levels[0]
scenarios = df_tot_emissions.columns.levels[1]
```

```python
col_names = ["Region", "Sector", "Year", "Scenario", "Emissions"]

df_emissions_pivot = pd.DataFrame(columns = ["Region", "Sector", "Year", "Scenario", "Emissions"])

for region in regions:
    for sector in sectors:
        for year in years:
            for scenario in scenarios:
                if (region, sector) in df_tot_emissions.index:
                    if (year, scenario) in df_tot_emissions.columns:
                        value = df_tot_emissions.loc[region,sector][year,scenario]
                        df_emissions_pivot = df_emissions_pivot.append(pd.DataFrame([[region, sector, year, scenario, value]],
                                                                       columns = ["Region", "Sector", "Year", "Scenario", "Emissions"]),
                                                                       ignore_index = True)
                
df_emissions_pivot
```

## Make pivot table from captured emissions

```python
# sectors are written differently in captured emissions

sectors_cap   = df_captured_emissions.index.levels[1]
```

```python
col_names = ["Region", "Sector", "Year", "Scenario", "Emissions"]

df_captured_pivot = pd.DataFrame(columns = ["Region", "Sector", "Year", "Scenario", "Emissions"])

for region in regions:
    for sector in sectors_cap:
        for year in years:
            for scenario in scenarios:
                if (region, sector) in df_captured_emissions.index:
                    if (year, scenario) in df_captured_emissions.columns:
                        value = df_captured_emissions.loc[region,sector][year,scenario]
                        df_captured_pivot = df_captured_pivot.append(pd.DataFrame([[region, sector, year, scenario, value]],
                                                                       columns = ["Region", "Sector", "Year", "Scenario", "Emissions"]),
                                                                       ignore_index = True)
                
df_captured_pivot
```

```python
with pd.ExcelWriter("emissions_TIMES_PanEU.xlsx") as writer:
    df_emissions_pivot.to_excel(writer, sheet_name="total_emissions", index = False)
    df_captured_pivot.to_excel(writer, sheet_name="captured_emissions", index = False)
```
